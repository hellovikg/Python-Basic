## Question 1: What is hypothesis testing in statistics?
Answer: Hypothesis testing is a statistical method used to determine if there is enough evidence in a sample data to draw conclusions about a population. It involves formulating a null hypothesis (H0) and an alternative hypothesis (H1), collecting data, and using statistical tests to assess the evidence against the null hypothesis.

## Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
Answer: The null hypothesis (H0) is a statement of no effect or no difference, often representing the status quo. The alternative hypothesis (H1) is a statement that contradicts the null hypothesis and represents what the researcher is trying to find evidence for. For example, if we are testing a new drug, the null hypothesis might be that the drug has no effect, while the alternative hypothesis might be that the drug has a positive effect.

## Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
Answer: The significance level (alpha or Î±) is the probability of rejecting the null hypothesis when it is actually true (Type I error). It is typically set at 0.05 (or 5%), meaning there is a 5% chance of incorrectly rejecting the null hypothesis. If the p-value of the statistical test is less than or equal to the significance level, we reject the null hypothesis.

## Question 4: What are Type I and Type II errors? Give examples of each.
Answer:
*   **Type I error (False Positive):** Rejecting the null hypothesis when it is true.
    *   *Example:* A medical test incorrectly diagnoses a healthy person with a disease.
*   **Type II error (False Negative):** Failing to reject the null hypothesis when it is false.
    *   *Example:* A medical test fails to detect a disease in a person who actually has it.

## Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.
Answer:
*   **Z-test:** Used when the population standard deviation is known and the sample size is large (typically n > 30). It assumes the data is normally distributed.
*   **T-test:** Used when the population standard deviation is unknown and the sample size is small (typically n < 30). It assumes the data is normally distributed.

## Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

n = 10  # number of trials
p = 0.5 # probability of success
size = 1000 # number of samples

binomial_data = np.random.binomial(n, p, size)

plt.hist(binomial_data, bins=range(n + 2), align='left', rwidth=0.8)
plt.xlabel("Number of successes")
plt.ylabel("Frequency")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()
```

## Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python.

```python
import numpy as np
from statsmodels.stats.weightstats import ztest

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Assume the population mean (mu) is 50 and population standard deviation (sigma) is known (e.g., 1.0 for demonstration)
# In a real scenario, sigma would be known or estimated from a large sample.
# For this example, let's perform a test against a hypothetical mean of 50.
mu = 50
sigma = 1.0 # Assuming known population standard deviation

# Perform Z-test
z_statistic, p_value = ztest(sample_data, value=mu, ddof=0)

print(f"Sample Mean: {np.mean(sample_data):.2f}")
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print(f"With a p-value of {p_value:.4f} and a significance level of {alpha}, we reject the null hypothesis.")
    print("Conclusion: There is significant evidence to suggest that the population mean is different from 50.")
else:
    print(f"With a p-value of {p_value:.4f} and a significance level of {alpha}, we fail to reject the null hypothesis.")
    print("Conclusion: There is not enough evidence to suggest that the population mean is different from 50.")
```

## Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Simulate data from a normal distribution
mu = 100  # mean
sigma = 15 # standard deviation
size = 200 # number of samples

normal_data = np.random.normal(mu, sigma, size)

# Calculate the 95% confidence interval for the mean
sample_mean = np.mean(normal_data)
sample_std = np.std(normal_data, ddof=1) # Use ddof=1 for sample standard deviation
n = len(normal_data)
alpha = 0.05
z_critical = norm.ppf(1 - alpha/2) # Z-score for 95% confidence

margin_of_error = z_critical * (sample_std / np.sqrt(n))
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval for the Mean: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")

# Plot the data
plt.hist(normal_data, bins=20, density=True, alpha=0.6, color='g')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)
plt.plot(x, p, 'k', linewidth=2)
title = "Fit results: mu = %.2f,  std = %.2f" % (mu, sigma)
plt.title(title)
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

## Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculates the Z-scores for a given dataset.

    Args:
        data (list or numpy array): The input dataset.

    Returns:
        numpy array: The Z-scores for the dataset.
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

# Example usage with a sample dataset
sample_data_z = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                  50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                  50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                  50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

z_scores = calculate_z_scores(sample_data_z)

print("Z-scores:", z_scores)

# Visualize the standardized data using a histogram
plt.hist(z_scores, bins=20)
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.title("Histogram of Z-scores")
plt.show()

print("\nExplanation of Z-scores:")
print("A Z-score represents how many standard deviations a data point is away from the mean of the dataset.")
print("A Z-score of 0 means the data point is equal to the mean.")
print("A positive Z-score means the data point is above the mean.")
print("A negative Z-score means the data point is below the mean.")
print("For example, a Z-score of 1.5 means the data point is 1.5 standard deviations above the mean.")
```
