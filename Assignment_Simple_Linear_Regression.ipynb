## Question 1: What is Simple Linear Regression?

**Answer:** Simple Linear Regression (SLR) is a statistical method used to model the relationship between two continuous variables: a dependent variable (or response variable) and an independent variable (or predictor variable). It assumes a linear relationship, meaning the change in the dependent variable is proportional to the change in the independent variable. The goal is to find the best-fitting straight line (the regression line) that minimizes the sum of squared differences between the observed and predicted values.

## Question 2: What are the key assumptions of Simple Linear Regression?

**Answer:** The key assumptions of Simple Linear Regression, often remembered by the acronym **LINE**, are:

1.  **Linearity:** The relationship between the independent and dependent variables is linear. This means the mean of the response variable is a straight-line function of the predictor variable.
2.  **Independence:** The observations are independent of each other. This is particularly important for time series data or repeated measures, where observations might be correlated.
3.  **Normality:** For any given value of X, the residuals (the differences between observed and predicted Y values) are normally distributed. This assumption is more critical for smaller sample sizes and for hypothesis testing.
4.  **Equal Variance (Homoscedasticity):** The variance of the residuals is constant across all levels of the independent variable. In other words, the spread of the residuals should be roughly the same throughout the range of X values.

Additional assumptions include:
*   **No measurement error in X:** The independent variable is measured without error.

## Question 3: What is heteroscedasticity, and why is it important to address in regression models?

**Answer:**

**Heteroscedasticity** (pronounced het-er-oh-skeh-DAS-tis-i-tee) is the violation of the equal variance assumption in regression. It occurs when the variance of the residuals (the errors or unexplained variance) is not constant across all levels of the independent variable(s). Visually, if you plot residuals against predicted values, you might see a funnel shape, where the spread of residuals increases or decreases as the predicted value changes.

**Why it is important to address:**

1.  **Inefficient Estimators:** While OLS (Ordinary Least Squares) estimators remain unbiased even with heteroscedasticity, they are no longer the most efficient (i.e., they don't have the smallest variance among all unbiased estimators). This means your standard errors might be larger than they should be, leading to less precise estimates of the regression coefficients.
2.  **Invalid Standard Errors and p-values:** The most significant problem is that heteroscedasticity causes the standard errors of the regression coefficients to be biased. If the standard errors are biased, then t-statistics and p-values will also be biased. This can lead to incorrect conclusions about the statistical significance of the predictors (e.g., falsely concluding a variable is significant or not significant).
3.  **Invalid Confidence Intervals:** Biased standard errors also lead to incorrect confidence intervals for the regression coefficients, making them unreliable.
4.  **Misleading Prediction Intervals:** Prediction intervals for new observations will also be inaccurate, as they rely on the assumption of constant error variance.

In summary, heteroscedasticity doesn't bias the regression coefficients themselves, but it invalidates the statistical tests and confidence intervals, making it difficult to draw reliable conclusions from the model.

## Question 4: What is Multiple Linear Regression?

**Answer:** Multiple Linear Regression (MLR) is an extension of Simple Linear Regression. Instead of using just one independent variable to predict a dependent variable, MLR uses two or more independent variables to predict the value of a single dependent (or response) variable. The model still assumes a linear relationship between the dependent variable and each independent variable. The goal is to find the best-fitting linear equation that describes how the dependent variable changes with respect to changes in the multiple independent variables, while minimizing the sum of squared residuals.

The general form of a multiple linear regression equation is:

`Y = β₀ + β₁X₁ + β₂X₂ + ... + βₚXₚ + ε`

Where:
*   `Y` is the dependent variable.
*   `β₀` is the y-intercept (the value of Y when all X's are zero).
*   `β₁, β₂, ..., βₚ` are the regression coefficients for each independent variable, representing the change in Y for a one-unit change in the respective X, holding all other X's constant.
*   `X₁, X₂, ..., Xₚ` are the independent variables.
*   `ε` is the error term (residuals).

## Question 5: What is polynomial regression, and how does it differ from linear regression?

**Answer:**

**Polynomial Regression** is a form of regression analysis in which the relationship between the independent variable (X) and the dependent variable (Y) is modeled as an nth-degree polynomial. While it models a curvilinear relationship, it is still considered a type of linear model because the equation is linear in its parameters (coefficients).

The general form of a polynomial regression equation of degree 'n' is:

`Y = β₀ + β₁X + β₂X² + ... + βₙXⁿ + ε`

**How it differs from Linear Regression:**

1.  **Nature of Relationship:**
    *   **Linear Regression:** Assumes a strictly linear relationship between X and Y. The regression line is a straight line.
    *   **Polynomial Regression:** Models a curvilinear relationship. It can capture non-linear patterns by introducing polynomial terms (X², X³, etc.) of the independent variable. The resulting regression line is a curve.

2.  **Equation Form:**
    *   **Linear Regression (Simple):** `Y = β₀ + β₁X + ε`
    *   **Polynomial Regression:** Includes higher-order terms of X, such as `X²`, `X³`, etc., in the equation, allowing for curves.

3.  **Flexibility:**
    *   **Linear Regression:** Less flexible, suitable only when the relationship is approximately straight.
    *   **Polynomial Regression:** More flexible, capable of fitting a wider range of shapes and curves, making it useful when the linear assumption is too restrictive.

4.  **Interpretability:**
    *   **Linear Regression:** Coefficients are generally easier to interpret (a one-unit change in X leads to a `β₁` change in Y).
    *   **Polynomial Regression:** Coefficients of higher-order terms are harder to interpret directly in terms of a simple linear change, as the effect of X on Y changes depending on the value of X itself.

## Question 6: Implement a Python program to fit a Simple Linear Regression model to the following sample data:

*   `X = [1, 2, 3, 4, 5]`
*   `Y = [2.1, 4.3, 6.1, 7.9, 10.2]`

Plot the regression line over the data points.

**Answer:**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshape X for sklearn
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create a Linear Regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, Y)

# Get the coefficients (slope and intercept)
intercept = model.intercept_
slope = model.coef_[0]

print(f"Intercept: {intercept:.2f}")
print(f"Slope: {slope:.2f}")

# Predict Y values using the fitted model
Y_pred = model.predict(X)

# Plot the original data points and the regression line
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='blue', label='Original Data Points')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.title('Simple Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

## Question 7: Fit a Multiple Linear Regression model on this sample data:

*   `Area = [1200, 1500, 1800, 2000]`
*   `Rooms = [2, 3, 3, 4]`
*   `Price = [250000, 300000, 320000, 370000]`

Check for multicollinearity using VIF and report the results.

**Answer:**

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Sample Data
Area = np.array([1200, 1500, 1800, 2000])
Rooms = np.array([2, 3, 3, 4])
Price = np.array([250000, 300000, 320000, 370000])

# Create a DataFrame for independent variables
X_df = pd.DataFrame({'Area': Area, 'Rooms': Rooms})
Y = Price

# Fit Multiple Linear Regression model
model_mlr = LinearRegression()
model_mlr.fit(X_df, Y)

print("Multiple Linear Regression Model Coefficients:")
print(f"Intercept: {model_mlr.intercept_:.2f}")
print(f"Coefficient for Area: {model_mlr.coef_[0]:.2f}")
print(f"Coefficient for Rooms: {model_mlr.coef_[1]:.2f}")

# Check for multicollinearity using VIF
# VIF is calculated for each independent variable
vif_data = pd.DataFrame()
vif_data['feature'] = X_df.columns
vif_data['VIF'] = [variance_inflation_factor(X_df.values, i) for i in range(X_df.shape[1])]

print("\nVariance Inflation Factor (VIF) Results:")
print(vif_data)
```

## Question 8: Implement polynomial regression on the following data:

*   `X = [1, 2, 3, 4, 5]`
*   `Y = [2.2, 4.8, 7.5, 11.2, 14.7]`

Fit a 2nd-degree polynomial and plot the resulting curve.

**Answer:**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshape X for sklearn
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Create polynomial features (degree 2)
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X)

# Fit a Linear Regression model to the polynomial features
model_poly = LinearRegression()
model_poly.fit(X_poly, Y)

# Predict Y values using the fitted polynomial model
Y_pred_poly = model_poly.predict(X_poly)

# For plotting the curve smoothly, create more points along the range of X
X_plot = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
X_plot_poly = poly_features.transform(X_plot)
Y_plot_pred = model_poly.predict(X_plot_poly)

print("Polynomial Regression Model Coefficients:")
print(f"Intercept: {model_poly.intercept_:.2f}")
print(f"Coefficients: {model_poly.coef_[1:]}") # Exclude the coefficient for X^0 (intercept)

# Plot the original data points and the polynomial regression curve
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='blue', label='Original Data Points')
plt.plot(X_plot, Y_plot_pred, color='red', label='2nd-Degree Polynomial Regression')
plt.title('Polynomial Regression (Degree 2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

## Question 9: Create a residuals plot for a regression model trained on this data:

*   `X = [10, 20, 30, 40, 50]`
*   `Y = [15, 35, 40, 50, 65]`

Assess heteroscedasticity by examining the spread of residuals.

**Answer:**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample Data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Fit a Simple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Predict Y values
Y_pred = model.predict(X)

# Calculate residuals
residuals = Y - Y_pred

# Create a residuals plot
plt.figure(figsize=(8, 6))
plt.scatter(Y_pred, residuals, color='blue')
plt.axhline(y=0, color='red', linestyle='--') # Add a horizontal line at 0
plt.title('Residuals Plot')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.grid(True)
plt.show()

print("Intercept: {:.2f}".format(model.intercept_))
print("Coefficient (Slope): {:.2f}".format(model.coef_[0]))
```

## Question 10: Imagine you are a data scientist working for a real estate company. You need to predict house prices using features like area, number of rooms, and location. However, you detect heteroscedasticity and multicollinearity in your regression model. Explain the steps you would take to address these issues and ensure a robust model.

**Answer:**

As a data scientist encountering heteroscedasticity and multicollinearity in a real estate pricing model, I would take the following steps to address these issues and build a robust model:

### Addressing Heteroscedasticity:

Heteroscedasticity violates the assumption of constant variance of residuals, leading to inefficient coefficient estimates and biased standard errors, which affects p-values and confidence intervals. Here's how I would address it:

1.  **Transform the Dependent Variable (Y):**
    *   **Log Transformation:** A common approach is to apply a logarithmic transformation to the dependent variable (e.g., `log(Price)`). This often helps to compress the range of the dependent variable and stabilize the variance. This is particularly useful when the variance increases with the mean.
    *   **Square Root Transformation:** For count data or moderately skewed data, a square root transformation (`sqrt(Price)`) might be effective.
    *   **Box-Cox Transformation:** This is a more general transformation that can automatically determine the optimal power transformation for the dependent variable to achieve normality and homoscedasticity.

2.  **Weighted Least Squares (WLS):** If the pattern of heteroscedasticity is known (e.g., variance is proportional to a certain independent variable), WLS can be used. This method assigns different weights to each observation based on the inverse of its error variance, giving less weight to observations with larger variance.

3.  **Robust Standard Errors:**
    *   Also known as White's standard errors or Huber-White standard errors. These standard errors adjust for heteroscedasticity without altering the regression coefficients themselves. They provide valid p-values and confidence intervals even in the presence of heteroscedasticity. This is often a practical first step if transformations don't significantly improve the issue or are difficult to interpret.

4.  **Re-evaluating Model Specification:**
    *   Sometimes, heteroscedasticity can indicate that important variables are missing from the model or that the functional form is incorrect (e.g., a linear model is being used for a non-linear relationship). I would explore adding interaction terms, polynomial terms, or other relevant features, or consider non-linear models if appropriate.

### Addressing Multicollinearity:

Multicollinearity occurs when independent variables in a regression model are highly correlated with each other, leading to unstable and unreliable coefficient estimates, large standard errors, and difficulty in interpreting the individual effect of predictors.

1.  **Identify Multicollinearity:**
    *   **Variance Inflation Factor (VIF):** Calculate VIF for each independent variable. A VIF value greater than 5 or 10 typically indicates significant multicollinearity.
    *   **Correlation Matrix:** Examine the correlation matrix of independent variables for high pairwise correlations.

2.  **Strategies to Mitigate Multicollinearity:**
    *   **Remove Highly Correlated Variables:** If two or more variables are highly correlated and conceptually similar (e.g., 'Area' and 'Number of Bedrooms' which might be highly correlated), I would consider removing one of them, choosing the one that is less theoretically important or has less predictive power.
    *   **Combine Variables (Feature Engineering):** Create a new composite variable from the highly correlated ones. For instance, instead of 'Area' and 'Number of Rooms', I might create 'Area per Room' or use principal component analysis (PCA).
    *   **Principal Component Analysis (PCA):** This technique transforms the original correlated features into a new set of uncorrelated features called principal components. These components can then be used in the regression model instead of the original features.
    *   **Ridge Regression or Lasso Regression:** These are regularization techniques that introduce a penalty term to the least squares objective function, which helps to shrink the regression coefficients. Ridge regression is effective in handling multicollinearity by reducing the variance of the estimates, while Lasso can also perform feature selection by shrinking some coefficients to zero.
    *   **Increase Sample Size:** While not always feasible, a larger sample size can sometimes alleviate multicollinearity by providing more diverse data points, leading to more stable estimates.

### Ensuring a Robust Model:

Beyond addressing these specific issues, I would also:

*   **Cross-Validation:** Use techniques like k-fold cross-validation to assess the model's performance on unseen data and ensure its generalizability.
*   **Outlier Detection and Treatment:** Identify and appropriately handle outliers, as they can disproportionately influence regression coefficients and residuals.
*   **Model Validation and Interpretation:** Continuously evaluate model fit using metrics like R-squared, Adjusted R-squared, MAE, RMSE, and ensure that the coefficients are logically interpretable in the context of real estate.
*   **Domain Expertise:** Collaborate with real estate experts to ensure that the chosen features and model assumptions align with practical knowledge of the market.
